## 2. Write a NumPy program to create an element-wise comparison (greater,    greater_equal, less and less_equal) of two given arrays.

```python
arr1 = np.array([10, 20, 30, 40, 50])
arr2 = np.array([20, 15, 40, 30, 50])
print("greater : ",np.greater(arr1, arr2))
print("Greater equal : ",np.greater_equal(arr1, arr2))
print("Less : ",np.less(arr1, arr2))
print("less equal : ",np.less_equal(arr1, arr2))
```

## 6. Write a NumPy program to get the element-wise remainder of an array of numbers from 20 to 50 which are divisible by 3.

```python
arr = np.arange(20, 51)
div_by3 = arr[arr % 3 == 0]
reminder = np.remainder(div_by3, 3)
print("Divisible by 3 :",div_by3)
print("Remainder: ",reminder)
```

## 7. Write a Pandas program to add, subtract, multiple and divide two Pandas Series.

```python
series1 = pd.Series([2, 4, 6, 8, 10])
series2 = pd.Series([1, 3, 5, 7, 9])
print("Addition : ", series1 + series2)
print("Substraction : ", series1 - series2)
print("Multiplication : ", series1 * series2)
print("Division : ", series1 / series2)
```

```python
df = pd.read_csv("D:/DS workshop/mtcar.csv")
print(df)
```

## i) Line Plot (mpg vs wt)

```python
plt.plot(df['wt'], df['mpg'])
plt.xlabel("Weight (wt)")
plt.ylabel("Miles per Gallon (mpg)")
plt.title("Line Plot: MPG vs Weight")
plt.show()
```

## Bar Graph

```python
df.groupby('cyl')['mpg'].mean().plot(kind='bar')
plt.xlabel("Cylinders")
plt.ylabel("Average MPG")
plt.title("Bar Graph: Cylinders vs Average MPG")
plt.show()
```

## iii) Pie Chart

```python
df['cyl'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Pie Chart: Cylinder Distribution")
plt.ylabel("")
plt.show()
```

## iv) Boxplot

```python
plt.boxplot(df['mpg'])
plt.title("Boxplot of MPG")
plt.ylabel("Miles per Gallon")
plt.show()
```

## v) Pairplot

```python
sns.pairplot(df[['mpg', 'cyl', 'disp', 'hp', 'wt']])
plt.show()
```

## vi) Scatter Plot

```python
plt.scatter(df['hp'], df['mpg'])
plt.xlabel("Horsepower (hp)")
plt.ylabel("Miles per Gallon (mpg)")
plt.title("Scatter Plot: MPG vs HP")
plt.show()
```

## vii) Histogram

```python
plt.hist(df['mpg'], bins=10)
plt.xlabel("Miles per Gallon")
plt.ylabel("Frequency")
plt.title("Histogram of MPG")
plt.show()
```
