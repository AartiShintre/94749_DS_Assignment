## xiv) 

```python
a) display records whose salary is greater than 50000.
greter_than_50k = df[df['salary']>50000]
print("Salary Greater than 50k",greter_than_50k)
```

```python
## b) display records whose salary is less than 20000.
less_than_20k = df[df['salary']<20000]
print("Salary less than 20k",less_than_20k)
```

```python
# c) display records whose salary is between 20000 to 50000.
between_20k_50k = df[(df['salary'] >= 20000) & (df['salary'] <= 50000)]
print("Salary between 20000 and 50000:\n", between_20k_50k)
```

```python
# d) Records whose salary is greater than 100000
greater_100k = df[df['salary'] > 100000]
print("Salary > 100000:\n", greater_100k)
```

```python
# e) Records whose salary is less than 100000
less_100k = df[df['salary'] < 100000]
print("Salary < 100000:\n", less_100k)
```

## xv)

```python
# a) Records whose discipline is 'A'
discipline_A = df[df['discipline'] == 'A']
print("Discipline = A:\n", discipline_A)

# b) Records whose discipline is 'B'
discipline_B = df[df['discipline'] == 'B']
print("Discipline = B:\n", discipline_B)

# c) Records whose service is greater than 10
service_gt_10 = df[df['service'] > 10]
print("Service > 10:\n", service_gt_10)

# d) Records whose service is less than 5
service_lt_5 = df[df['service'] < 5]
print("Service < 5:\n", service_lt_5)
```
