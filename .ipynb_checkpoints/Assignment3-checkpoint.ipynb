## Following is the description of columns in dataset ToyotaCorolla.csv. Prepare a prediction model for predicting Price

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings(action="ignore")
```

```python
df=pd.read_csv("ToyotaCorolla.csv",encoding='latin-1')
df.head()
```

```python
df.info()
```

```python
# get the statistical information about the data
df.describe()
```

```python
# get a quick view of data
df.head()
```

```python
# get a quick view of last few records
df.tail()
```

```python
# check if df has any NA value
df.isna().sum()
```

```python
# create x
# set of independent variables
#- is is mandatory to get the x as 2D array or data frame
#- axis: 0-> row, 1-> column
#- df.drop() will remove the column and return a new dataframe object
#- it does not update the existing data frame
x = df[["Age_08_04","KM","HP","cc","Doors","Gears","Quarterly_Tax","Weight"]]
#create y
#- dependent variable
#- it is mandatory to get the y as 1D array or series
y = df['Price']
```

```python
df.head()
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
# split the data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=123456)
```

```python
from sklearn.linear_model import LinearRegression
# create a model
model = LinearRegression()
# train the model using the x and y
model.fit(x, y)
```

```python
new_car = [[24,45000,110,1600,4,5,180,1250]]
prediction = model.predict(new_car)
print(f"Predicted Price= {prediction[0]:.2f}")
```

```python
# predict the profit values for test set
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
# mean absolute error (MAE)
mae = mean_absolute_error(y_test, y_pred)
mae
```

```python
from sklearn.metrics import mean_squared_error
# mean squared error
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```

```python
from sklearn.metrics import r2_score
# get r2 score
r2 = r2_score(y_test, y_pred)
r2
```

```python
plt.figure()
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Price")
plt.show()
```

```python
model.score(x,y)
```
