```python
df['vehicle_volume'] = df['Length'] * df['width'] * df['height']
df['weight_to_volume'] = df['Gross_weight'] / (df['vehicle_volume'] + 0.01)
df['length_width_ratio'] = df['Length'] / (df['width'] + 0.01)
df['height_width_ratio'] = df['height'] / (df['width'] + 0.01)
df['height_length_ratio'] = df['height'] / (df['Length'] + 0.01)
df['weight_per_cylinder'] = df['Gross_weight'] / (df['cylinder'] + 0.01)
df['weight_per_airbag'] = df['Gross_weight'] / (df['airbags'] + 0.01)
df['volume_per_cylinder'] = df['vehicle_volume'] / (df['cylinder'] + 0.01)
df['density_index'] = df['Gross_weight'] / (df['Length'] + 0.01)
df['compactness_score'] = df['Gross_weight'] / (df['width'] + 0.01)

```

```python
df['safety_score'] = (
    df['airbags'] +
    df['is_esc'] +
    df['is_tpms'] +
    df['is_parking_sensors'] +
    df['is_parking_camera']
)

df['safety_per_weight'] = df['safety_score'] / (df['Gross_weight'] + 0.01)
df['airbag_density'] = df['airbags'] / (df['vehicle_volume'] + 0.01)
df['safety_per_cylinder'] = df['safety_score'] / (df['cylinder'] + 0.01)
df['ncap_weight_ratio'] = df['ncap_rating'] / (df['Gross_weight'] + 0.01)
df['ncap_engine_ratio'] = df['ncap_rating'] / (df['displacement'] + 0.01)
df['ncap_safety_ratio'] = df['ncap_rating'] / (df['safety_score'] + 0.01)
```

```python
df['density_weight_ratio'] = df['population_density'] / (df['Gross_weight'] + 0.01)
df['density_engine_ratio'] = df['population_density'] / (df['displacement'] + 0.01)
df['density_volume_ratio'] = df['population_density'] / (df['vehicle_volume'] + 0.01)
df['turning_weight_ratio'] = df['turning_radius'] / (df['Gross_weight'] + 0.01)
df['turning_length_ratio'] = df['turning_radius'] / (df['Length'] + 0.01)
```

```python


df['policy_vehicle_ratio'] = df['policy_tenure'] / (df['age_of_car'] + 0.01)
df['policy_owner_ratio'] = df['policy_tenure'] / (df['age_of_policyholder'] + 0.01)
df['car_owner_age_gap'] = df['age_of_policyholder'] - df['age_of_car']
df['policy_age_product'] = df['policy_tenure'] * df['age_of_car']
df['owner_policy_sum'] = df['age_of_policyholder'] + df['policy_tenure']
df['owner_policy_diff'] = df['age_of_policyholder'] - df['policy_tenure']
df['vehicle_life_fraction'] = df['age_of_car'] / (df['age_of_policyholder'] + 0.01)
df['policy_square'] = df['policy_tenure'] ** 2
```

```python
df['displacement_per_gear'] = df['displacement'] / (df['gear_box'] + 0.01)
df['cylinder_per_gear'] = df['cylinder'] / (df['gear_box'] + 0.01)
df['engine_stress_index'] = df['displacement'] * df['cylinder']
df['engine_weight_ratio'] = df['displacement'] / (df['Gross_weight'] + 0.01)
df['cylinder_weight_ratio'] = df['cylinder'] / (df['Gross_weight'] + 0.01)
df['gear_weight_ratio'] = df['gear_box'] / (df['Gross_weight'] + 0.01)
df['turning_per_gear'] = df['turning_radius'] / (df['gear_box'] + 0.01)
df['engine_density_ratio'] = df['displacement'] / (df['Length'] + 0.01)
```

```python
df['base_area'] = df['Length'] * df['width']
df['side_area'] = df['Length'] * df['height']
df['front_area'] = df['width'] * df['height']
df['dimension_sum'] = df['Length'] + df['width'] + df['height']
df['dimension_mean'] = (df['Length'] + df['width'] + df['height']) / 3
df['weight_per_length'] = df['Gross_weight'] / (df['Length'] + 0.01)
df['weight_per_height'] = df['Gross_weight'] / (df['height'] + 0.01)
```

```python
df['active_safety_count'] = (
    df['is_esc'] +
    df['is_tpms'] +
    df['is_parking_sensors'] +
    df['is_parking_camera']
)

df['comfort_safety_score'] = (
    df['is_adjustable_steering'] +
    df['is_esc'] +
    df['is_tpms']
)

df['safety_to_weight'] = df['active_safety_count'] / (df['Gross_weight'] + 0.01)
df['airbags_per_weight'] = df['airbags'] / (df['Gross_weight'] + 0.01)
df['airbags_per_cylinder'] = df['airbags'] / (df['cylinder'] + 0.01)
df['ncap_per_airbag'] = df['ncap_rating'] / (df['airbags'] + 0.01)
df['ncap_per_weight'] = df['ncap_rating'] / (df['Gross_weight'] + 0.01)
```

```python
df['max_torque'] = df['max_torque'].astype(str)
df['max_power'] = df['max_power'].astype(str)

df['rpm'] = df['max_torque'].str.extract(r'(\d+)').astype(float)
df['torque_val'] = df['max_torque'].str.extract(r'(\d+)').astype(float)
df['power_val'] = df['max_power'].str.extract(r'(\d+)').astype(float)

df['torque_to_rpm_ratio'] = df['torque_val'] / df['rpm']
df['power_to_rpm_ratio'] = df['power_val'] / df['rpm']

```

```python
# Imports
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import SMOTE

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Apply SMOTE
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Scaling
scaler = StandardScaler()
X_train_smote = scaler.fit_transform(X_train_smote)
X_test = scaler.transform(X_test)

# Logistic Regression
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train_smote, y_train_smote)
print("Logistic Regression Accuracy:", accuracy_score(y_test, lr.predict(X_test)))

# KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_smote, y_train_smote)
print("KNN Accuracy:", accuracy_score(y_test, knn.predict(X_test)))

# SVM
svm = SVC()
svm.fit(X_train_smote, y_train_smote)
print("SVM Accuracy:", accuracy_score(y_test, svm.predict(X_test)))
```
