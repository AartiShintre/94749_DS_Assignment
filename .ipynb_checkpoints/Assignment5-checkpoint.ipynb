```python
# import required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 3. convert Categorical columns into numerical columns.

```python
df[['Surname','Geography','Gender']].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()

df['Surname'] = lb.fit_transform(df['Surname'])
df['Geography'] = lb.fit_transform(df['Geography'])
df['Gender'] = lb.fit_transform(df['Gender'])

df
```

```python
df.info()
```

```python
df.corr()
```

## 2. Drop Unnecessary Columns.

```python
df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1, inplace=True)
```

## KNN model building

```python
x = df.drop('Exited',axis=1)
y = df['Exited']
```

```python
# split the data into train and test sets
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=123456)
```

```python
 # K nearest neighbour
from sklearn.neighbors import KNeighborsClassifier

# create a model
model = KNeighborsClassifier(n_neighbors=5)

# fit the data
model.fit(x_train,y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
knn = KNeighborsClassifier()
param_grid={
    'n_neighbors':[3,5,7,9]
}

# setup the Grid Search
knn_cv = GridSearchCV(knn,param_grid,cv=5)

knn_cv.fit(x_train,y_train)
```

```python
# find the best parameters
print("Best parameter :",knn_cv.best_params_)
```

```python
# find the best validation score
print("Best cross validation score :",knn_cv.best_score_)
```

```python
y_pred = model.predict(x_test)
y_true = y_test
```

```python
# # Evaluate a model
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,roc_auc_score

accuracy = accuracy_score(y_true,y_pred)
precision = precision_score(y_true,y_pred)
recall = recall_score(y_true,y_pred)
f1 = f1_score(y_true,y_pred)
roc_auc_score = roc_auc_score(y_true,y_pred)

print(confusion_matrix(y_true,y_pred))
print(f"accuracy = {accuracy}")
print(f"precision = {precision}")
print(f"recall = {recall}")
print(f"f1 = {f1}")
print(f"roc_auc_score = {roc_auc_score}")
```

```python
plt.scatter(
    x_test['Age'][y_pred == 1],
    x_test['Balance'][y_pred == 1],
    color="green",
    label="Predicted = 1 (Exited)"
)

plt.scatter(
    x_test['Age'][y_pred == 0],
    x_test['Balance'][y_pred == 0],
    color="blue",
    label="Predicted = 0 (Not Exited)"
)

plt.xlabel("Age")
plt.ylabel("Balance")
plt.title("KNN PREDICTED OUTPUT")
plt.legend()
plt.show()
```

```python
plt.scatter(x_test['Age'][y_true==1],x_test['Balance'][y_true==1],color="green")
plt.scatter(x_test['Age'][y_true==0],x_test['Balance'][y_true==0],color="blue")
plt.xlabel("Age")
plt.ylabel("Balance")
plt.title("OBSERVED OUTPUT")
```

```python
X = df.drop('type', axis=1)
y = df['type']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size=0.8, random_state=123456
)
```

```python
# K nearest neighbour
from sklearn.neighbors import KNeighborsClassifier

# create a model
model = KNeighborsClassifier(n_neighbors=5)

# fit the data
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_pred = model.predict(X_test)
y_true = y_test
```

```python
from sklearn.metrics import (
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:\n", cm)

# Metrics (multiclass)
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='weighted')
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')

print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1 Score :", f1)

```
