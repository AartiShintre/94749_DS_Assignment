## for SVM

```python
# split the data
```

```python
x = df.drop(['Exited'],axis=1)
y = df['Exited']
```

```python

from sklearn.model_selection import train_test_split

#split the data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=123456)
```

```python
## for SVM
```

```python
from sklearn.svm import SVC
model_svc = SVC(C=2,kernel='rbf')
model_svc = (x_train,y_train)
```

```python
y_pred = grid_search_svm.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
print(f"accuracy = {accuracy:.2f}, precision = {precision:.2f}, recall = {recall:.2f}, f1 = {f1:.2f}")
```

```python
grid_search_svm.score(x,y)
```

## for Decision tree

```python
## Split the data
```

```python
x = df.drop(['Exited'],axis=1)
y = df['Exited']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=123456)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion='gini',max_depth=500)
model.fit(x_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV

#hyper-parameters
parameters = {
   "criterion": ["gini", "log_loss", "entropy"],
   "max_depth": np.arange(10) * 100
}

#create a grid search cv object
grid_search_cv = GridSearchCV(estimator=model, param_grid=parameters)

#fit the data
grid_search_cv.fit(x_train,y_train)
```

```python
grid_search_cv.best_params_
```

```python
y_pred = model.predict(x_test)
y_true = y_test
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve

cm = confusion_matrix(y_true, y_pred)
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)
roc = roc_auc_score(y_true, y_pred)
```

```python
print(f"accuracy = {accuracy * 100:0.2f}%")
print(f"precision = {precision * 100:0.2f}")
print(f"recall = {recall * 100:0.2f}")
print(f"f1 score = {f1:0.2f}")
print(f"roc score = {roc * 100:0.2f}%")
```

```python
plt.scatter(
        x_test['IsActiveMember'][y_pred==0], 
        x_test['Balance'][y_pred==0], 
        label="Not Exited",
        color="green")
plt.scatter(
        x_test['IsActiveMember'][y_pred==1], 
        x_test['Balance'][y_pred==1], 
        label="Exited",
        color="red")
plt.xlabel('IsActiveMember')
plt.ylabel('Balance')
plt.title('decision tree')
plt.legend()
```
